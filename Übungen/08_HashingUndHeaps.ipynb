## Hashing und Heaps

### Hashing
Diese Übung befasst sich mit dem Thema Hashing.

Was ist das und wofür brauch ich das?    
Das Hashing ist eine praktische Alternative zu herkömmlichen Suchmethoden. Es bietet eine effiziente Implementierung für dynamischer Mengen, insbesondere für Wörterbuchoperationen wie insert, search und delete. Da keine Vorarbeit wie z.B. Sorteieren der Daten nötig ist, ist es ein effektives Verfahren zum Einfügen und Suchen von Daten.


Wie funktioniert das?    
Mittels einer Hashfunktion werden die Daten an einer bestimmten Stellle in einer Hashtabelle gespeichert. Sucht man jetzt einen Datensatz, gibt man ihn einfach ihn die Hashfunktion ein und erhält so den Tabellenplatz, an dem der Datensatz gespeichert ist.
