# FMA Übungen
## Modellierung
In dieser Übung geht es um ein grundsätzliches Verständnis dafür, wie man ein Problem visualisieren und seinen Lösungsweg modellieren kann. Die folgenden Aufgaben müssen so gelöst werden, dass eine andere Person ohne Erklärung den Lösungsweg nachvollziehen kann. Dazu eignen sich am besten Skizzen.

### Grundsätze der Problemlösung
1. **Make a wish!**  
Wenn ich mir etwas wünschen dürfte, was wäre das?  
Kann ich tricksen, um das zu bekommen?  
2. **Get your hands dirty!**  
Einfach niederschreiben, bevor man zu lange einfach nur über eine elegante Lösung nachdenkt.  
3. **Nicht-Wissen ist Wissen!**  
Wenn ich weiß, dass ich etwas nicht weiß, weiß ich etwas.  
Mit diesem Wissen können unter Umständen Fälle ausgeschlossen werden.  
4. **Vertraue nie deiner Intuition!**  
Sie ist trügerisch und führt nicht unbedingt zur effizientesten Lösung.

### Game of Life  
Die Regeln:  
* Eine tote Zelle mit genau drei lebenden Nachbarn wird in der Folgegeneration neu geboren.
* Lebende Zellen mit weniger als zwei lebenden Nachbarn sterben in der Folgegeneration an Einsamkeit.
* Eine lebende Zelle mit zwei oder drei lebenden Nachbarn bleibt in der Folgegeneration am Leben.
* Lebende Zellen mit mehr als drei lebenden Nachbarn sterben in der Folgegeneration an Überbevölkerung.  
Die Aufgabenstellung:  
Gegeben ist eine Kollonie der Generation 0. Zeigen Sie, wie die Kollonie in der ersten und zweiten Kollonie ausehen wird.

### Kugeln in der Waage (schwer)
* Wir teilen die Kugeln in drei Päckchen zu je vier Kugeln auf und vergleichen das Gewicht zweier Päckchen miteinander.  
    * **Die Waage ist im Gleichgewicht:**  
    Ist die Waage bei der ersten Messung im Gleichgewicht nehmen wir die verbleibenden vier Kugeln, legen eine der Kugeln beiseite und vergleichen die restlichen drei Kugeln mit drei Referenzkugeln.  
        * **Die Waage ist im Gleichgewicht:**  
        Ist die Waage bei der zweiten Messung im Gleichgewicht brauchen wir die zur Seite gelegte Kugel nur noch mit einer Referenzkugel vergleichen.  
        * **Die Waage ist nicht im Gleichgewicht:**  
        Ist die Waage bei der zweiten Messung nicht im Gleichgewicht, wissen wir zumindest schonmal, ob die gesuchte Kugel leichter oder schwerer als die restlichen Kugeln ist. Nun müssen wir noch herausfinden, um welche Kugel es sich handelt. Dazu nehmen wir eine der drei Kugeln, sowie eine Referenzkugel von der Waage und vertauschen eine weitere Kugel mit einer Referenzkugel. Daraus wie sich die Stellung der Waage (verglichen mit der vorherigen Messung) verändert (oder auch nicht) können wir eindeutig darauf schließen, welche Kugel anders ist, als die anderen.  
    * **Die Waage ist nicht im Gleichgewicht:**  
    Ist die Waage bei der ersten Messung nicht im Gleichgewicht, nehmen wir auf einer Seite drei Kugeln von der Waage, legen von der anderen Seite drei Kugeln auf diese Seite und füllen die leeren Positionen mit drei Referenzkugeln auf. 
        * **Die Waage ist im Gleichgewicht:**  
        Ist die Waage bei der zweiten Messung im Gleichgewicht befindet sich die gesuchte Kugel unter den Kugeln, die wir von der Waage genommen haben und wir wissen bereits, ob die gesuchte Kugel schwerer oder leichter als die anderen ist. Wiegt man nun zwei dieser drei Kugeln gegeneinander ab, kann man bestimmen, welche die gesuchte Kugel ist.  
        * **Die Waage ist nicht im Gleichgewicht, die Position bleibt dieselbe:**  
        Ändert sich die Position bei der zweiten Messung nicht, handelt es sich um eine der beiden Kugeln, deren Position wir nicht verändert haben. Da wir hier aber ebenfalls bereits jeder Kugel eine Eigenschaft zuordnen können, müssen wir nur noch herausfinden, um welche Kugel es sich handelt. Dies tun wir, indem wir eine der beiden Kugeln gegen eine Referenzkugel abwiegen.
        * **Die Waage ist nicht im Gleichgewicht, die Position wechselt:**  
        Kippt die Waage bei der zweiten Messung auf die andere Seite, muss es eine der drei Kugeln sein, die wir von der einen auf die andere Seite umgelegt haben. Diesen drei Kugeln können wir nun bereits die Eigenschaft schwerer oder leichter zuordnen. Wir wiegen nun zwei der drei Kugeln gegeneinander ab. Ist die Waage im Gleichgewicht handelt es sich um die dritte Kugel. Ist sie es nicht, muss es die Kugel sein, welche die bereits bestimmte Eigenschaft besitzt. Aus der ersten Messung wissen wir aber schon, ob diese beiden Kugeln leichter sind als der Rest, oder nicht. Wir müssen nun von den beiden Kugeln nur noch je eine Kugel nehmen und mit einer Referenzkugel vergleichen um zu entscheiden, um welche der beiden Kugeln es sich handelt.

### Zahelenrätsel 
A2 ist die Lüge: 23 ist eine Primzahl, das Produkt müsste also aus 1·23 bestehen. 23 ist aber keine Ziffer.
   
B2 ist die Lüge: B2 und C1 können nicht gleichzeitig wahr sein, da kein dreistelliges Vielfaches von 11 aus drei gleichen Ziffern besteht.  
B2 und C2 können auch nicht gleichzeitig wahr sein, da die gesuchte Zahl sonst aus drei Nullen bestehen würde.  
B2 kann also nicht wahr sein.

Ist C1 wahr kommen die Zahlen 11·37=407 und 2·11·37=814 in Frage. (dreistellig, durch 37 und 11 teilbar)   
Ist C2 wahr kommen die Zahlen 10·37=370 und 20·37=740 in Frage. (dreistellig, durch 37 teilbar, endet auf 0)  

Ist D1 wahr kommen noch die Zahlen 740 und 814 in Frage.   
Ist D2 wahr kommt nur 370 in Frage.  

407 kommt nicht in Frage, da dann sowohl D1 als auch D2 wahr wären.

Die drei möglichen Zahlen auf dem Zettel sind also 370, 740 und 814.
